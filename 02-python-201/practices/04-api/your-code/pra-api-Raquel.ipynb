Queremos conocer los nombres más frecuentes de los recién nacidos en Barcelona por sexo entre los años 1996 y 2018. Usa la función para obtener lo siguiente:

a) Una lista con el nombre más frecuente de los recién nacidos en Barcelona por sexo para cada uno de los años entre 1996 y 2018 (incluidos).

Para realizar el ejercicio consultad el portal de datos abiertos del Ayuntamiento de Barcelona mediante la siguiente url.

```python
#import pandas as pd
#from scipy import stats
#import numpy as np
#import matplotlib.pyplot as plt
#import seaborn as sns
#import warnings
#warnings.filterwarnings("ignore")
#%matplotlib notebook
```

```python
# leemos el fichero con los nombres de los niños nacidos entre 1996 y 2019
bcn = pd.read_csv("C:/Users/raque/Downloads/25_noms_padro_any_sexe_1996_2019.csv", delimiter =",")
```

```python
#elimino los niños nacidos en 2019:
bcn2 = bcn[bcn["Any"] != 2019]

#eliminamos los espacios de la columna Nom:
bcn2["Nom"] = bcn2["Nom"].map(str.strip)

#agrupamos los nombres de los recien nacidos por año y por sexo,
#sumando en la columna Nombre el número de veces puesto cada año:

bcn3 = bcn2.groupby(["Any","Sexe","Nom"]).sum().sort_values(["Any","Sexe","Nombre"], ascending=False)

#seleccionamos los nombres más veces puestos entre 1996 y 2018 por sexo:
bcn4= bcn3.loc[bcn3.groupby(["Any","Sexe"])["Nombre"].idxmax()]
bcn4
```

```python
#b) Crear una lista ordenada de mayor a menor, a partir de la lista obtenida en el apartado a),
#según el número de veces que dicho nombre ha sido utilizado:
bcn5 = bcn4.groupby(["Sexe","Nom"]).sum().sort_values(["Sexe","Nombre"], ascending=False)
bcn5
#Los nombres más veces puestos para chico son Marc y Pol y para chica son María y Martina. 
```
