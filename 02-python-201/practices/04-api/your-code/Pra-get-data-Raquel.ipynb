***Respuesta***

Descripción de los principales códigos de error http:

200: Es una respuesta satisfactoria. La solicitud ha tenido éxito. 

301: Significa Moved Permanently, que implica que la URI del recurso solicitado ha sido cambiado. 

400: Bad request. Significa que la sintaxis es inválida y el servidor no ha podido interpretar la solicitud dada.

401: Unauthorized. Es necesario autenticar para poder obtener la respuesta solicitada. 

403: Forbidden. No se poseen los permisos necesarios para cierto contenido, por lo que el servidor rechaza la petición. 

404: Not found. El servidor no pudo encontrar el contenido solicitado. 

505: HTTP Version Not Supported. La versión de HTTP usada en la petición no está soportada por el servidor. 

501: Not implemented. Significa que el método solicitado no está soportado en el servidor y no puede ser manejado.

```python
#Creo una lista con las URLs de los ids facilitados:
listas = ["https://cat-fact.herokuapp.com/facts/5d38bdab0f1c57001592f156", 
          "https://cat-fact.herokuapp.com/facts/ed11e643c15f700172e3856", 
          "https://cat-fact.herokuapp.com/facts/5ef556dff61f300017030d4c",
          "https://cat-fact.herokuapp.com/facts/5d9d4ae168a764001553b388"]

#Creo una función llamada gatos que recorra dicha lista, convierta el texto en 
#un diccionario y luego haga un data frame e imprima
#una tabla con los resultados
def gatos():
    for lista in listas:
        response = requests.get(lista)
        response2 = json.loads(response.text)
        df = pd.DataFrame([[key, response2[key]] for key in response2.keys()],
                         columns = ["Key", "Description"])
        print("Tabla de ", lista,"\n")
        print(df)
        print("\n")
gatos()
```

Ejecuta las dos funciones sobre el usuario de twitter Space_Station.

```python
#Para la primera parte del ejercicio importamos las librerías tweepy y abrimos
#el fichero txt con las credenciales 

import tweepy
import sys, os
tw_creeds = open("creds.txt","r")
lines = tw_creeds.readlines()

#creo una función para que las lea y poder acceder a la API de Twitter:
def get_creds(line):
    keys = [] 
    for l in line:
        keys.append(l.split("=")[1].splitlines(False)[0]) 
    return keys

CONSUMER_KEY = get_creds(lines)[0] 
CONSUMER_SECRET = get_creds(lines)[1]
ACCESS_TOKEN = get_creds(lines)[2]
ACCESS_TOKEN_SECRET = get_creds(lines)[3]

auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)

#Lanzo la API:
api = tweepy.API(auth)

#Y obtengo la tupla del usuario, he cogido el usuario de la escritora
#@LaEtxebarria porque es el primero que me ha salido:

tweepy.models.User("@LaEtxebarria")
```

```python
#Para la segunda parte del ejercicio, defino el usuario:
user = api.get_user("@LaEtxebarria")

#Y creo una función que muestre los resultados del usuario dado:
# el numero de tweets, el de amigos, el de seguidores y los nombres y descripciones
#de algunos de sus amigos:
def twit(user):
    print("Los resultados del usuario @",user.screen_name,"son los siguientes:\n")
    print("El numero de tweets es: {}".format(user.statuses_count))
    print("El numero de amigos es: {}".format(user.friends_count))
    print("El numero de seguidores es: {}".format(user.followers_count))
    print("Los nombres y descripciones de sus amigos son: \n")
    for friend in user.friends():
        print(friend.screen_name)
        print(friend.description)
        print("\n")
   

twit(user)
```
