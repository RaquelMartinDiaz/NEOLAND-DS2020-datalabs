Añadid una nueva columna al _dataframe_ con un valor booleano indicando una predicción básica de si en una cierta batalla el bando atacante será vencedor. Definiremos el valor como _True_ sí el bando atacante es mayor al defensor, y como _False_ en caso contrario. Definid como _NaN_ aquellos casos en los que no haya número de tropas en ningún bando. ¿En que casos, identificados por el número de batalla (*battle\_number*), nuestra predicción coincide con el resultado real?

# Respuesta
Lo primero que voy a hacer es descargar la librería numpy. Después voy a reemplazar los valores NaN de las columnas attacker_size y defender_size por 0. 


```python
import numpy as np
```

```python
got["attacker_size"].fillna(0, inplace= True)
got["defender_size"].fillna(0, inplace= True)
```

Ahora, usando la librería numpy, creo una lista de condiciones y una lista de elecciones. Luego con select directamente elige cada elección según la condición de esa fila

```python
conditionlist = [
    (got["attacker_size"]>got["defender_size"]),
    (got["attacker_size"]<got["defender_size"]),
    (got["attacker_size"] == None) & (got["defender_size"] == None)
]
choicelist = ["True", "False", "NaN"]
got["winner"] = np.select(conditionlist, choicelist, default ="NaN")
```
